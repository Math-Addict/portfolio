## **U Food Marketing Analysis** 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
food = pd.read_csv(r"D:\data analysis\pandas\inpedth\u_food_marketing.csv")
food.head()
```

```python
pd.set_option('display.max.columns',50)
```

```python
food.head()
```

```python
#first lets check for duplicates
food[food.duplicated()]
```

```python
#drop duplicates
food.drop_duplicates(inplace = True)
```

```python
#checking if that work 
food[food.duplicated()]
```

```python
#check the info of the data frame
food.info()
```

```python
#lets add a column of total children 
food['Total_Children'] = food[['Kidhome','Teenhome']].sum(axis = 1)
```

```python
food.head()
```

```python
#lets add a marital column so we will be able to filter on that easily, we say 1-widowed,2-together,3-single,4-married,5-divorced
subset_columns = ['marital_Divorced','marital_Married','marital_Single','marital_Together','marital_Widow']
food['Marital_Status'] = food[subset_columns].idxmax(axis = 1) #return a column with the column names where we have 1
```

```python
#lets map the column names to the corresponding values
mapping = {'marital_Divorced': 5,'marital_Married': 4,'marital_Single': 3,'marital_Together': 2,'marital_Widow': 1}
food['Marital_Status'] = food['Marital_Status'].map(mapping)
```

```python
food.head()
```

```python
#lets add another column of marital so we can be able to group on that if we want
marital = {5:'Divorced',4: 'Married',3: 'Single',2: 'Together',1: 'Widow'}
food['Marital_status_str'] = food['Marital_Status'].map(marital)
```

```python
#lets do the same thing for education
education_columns = ['education_2n Cycle','education_Basic','education_Graduation','education_Master','education_PhD']
food['Education_Status'] = food[education_columns].idxmax(axis = 1)
education_mapping = {'education_2n Cycle': 1,'education_Basic': 2,'education_Graduation': 3,'education_Master': 4,'education_PhD': 5}
food['Education_Status'] = food['Education_Status'].map(education_mapping)
food.head()
```

```python
#lets make a column that is 1 if they accepted any campaign and 0 else
food['Accepted_Campaigns'] = (food['AcceptedCmpOverall'] != 0).astype(int)
```

```python
#lets find the corr between the accepted campaigns and the other columns
food.corr(numeric_only = True)['Accepted_Campaigns'].sort_values(ascending = False)
```

```python
#lets try and visualize that but first lets filter a little
all_correlations = food.corr(numeric_only = True)
all_correlations = all_correlations[(all_correlations > 0.3) & (all_correlations < 1)]
sns.heatmap(all_correlations)
```

```python
#now lets add a column of age groups 
bins = [0,20,30,40,50,60,70,85,float('inf')]
labels =['<20','20-30', '31-40', '41-50', '51-60', '61-70', '71-85','85+']
food['Age_Group'] = pd.cut(food['Age'],bins = bins,labels = labels)
```

```python
food.head()
```

```python
#lets visualize the age group against campaigns
sns.pointplot(data = food,x = 'Age_Group',y = 'AcceptedCmpOverall')
```

```python
#lets see how much people in each age group
group_counts = food['Age_Group'].value_counts()
```

```python
#lets make that in percentage
percentage = group_counts/food.shape[0]*100
percentage
```

```python
percent_food = percentage.reset_index()
```

```python
percent_food.columns = ['age_group','percentage']
```

```python
percent_food = percent_food.sort_values('age_group')
```

```python
percent_food
```

```python
#lets visualize that
sns.barplot(x = 'age_group',y = 'percentage',data = percent_food ,hue = 'age_group')
plt.title('Percentage of Accepted Campaigns per Age Group')
plt.show()
```

```python
# Age segmentation - core audience for accepting campaigns right now is 31-70
```

```python
#lets check how much each group age is spending
spending = food.groupby('Age_Group',observed = True)['MntTotal'].sum().reset_index()
```

```python
sns.barplot(x = 'Age_Group',y = 'MntTotal',data = spending ,hue = 'Age_Group')
plt.title('Amount Spent Per Age Group')
plt.show()
```

```python
#now lets do the same just filter at the accepted campaigns
acct_cam = food[food['Accepted_Campaigns'] != 0] 

acceptence = acct_cam.groupby('Age_Group',observed = True)['MntTotal'].sum().reset_index()

sns.barplot(x = 'Age_Group',y = 'MntTotal',data = acceptence ,hue = 'Age_Group')
plt.title('Amount Spent Per Age Group When Accepted Campaign')
plt.show()
```
