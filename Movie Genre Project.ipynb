```python
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read the data into a data frame
movies = pd.read_csv(r"D:\data analysis\pandas\inpedth\imdb_movies.csv")
movies.head()
```

```python
movies.head()
```

```python
#lets check for duplicated
movies[movies.duplicated()]
```

```python
#we saw that we have 1 row that is duplicate so lets remove her
movies.drop_duplicates(inplace = True)
```

```python
movies[movies.duplicated()]
```

```python
#we need to populate the genre column so we can answer our questions
movies['genres'].isnull().sum()
```

```python
#we can see we have 23 null values in the genre column
#we will remove that null values because it will mess with our data and our questions
movies.dropna(subset = ['genres'], inplace = True)
```

```python
movies['genres'].isnull().sum() #checking if there is no null anymore
```

```python
#lets add a column off profit
movies['profit'] = movies['revenue'] - movies['budget']
```

```python
movies.head()
```

```python
#now lets drop some columns we dont need for what we want to check.
column_to_keep = ['popularity','budget','revenue','original_title','runtime','genres','release_year','vote_count','vote_average','profit']
movies_genres = movies[column_to_keep]
```

```python
#lets split the genre column up
split = movies_genres['genres'].str.split('|').apply(Series,1).stack() #split the genres and then convert them into a panda series and then stack everything 
split.index = split.index.droplevel(-1) #drops the inner most index level
split
```

```python
#for the sake of memory we will delete the genre column and replace it by our new one
split.name = 'genres_split'
del movies_genres['genres']
movies_genres = movies_genres.join(split)
```

```python
movies_genres.head()
```

1. Which genres are the most commmon (number of movies made)?

```python

genres_count = movies_genres.groupby('genres_split')[['original_title']].nunique().sort_values(by ='original_title')
```

```python
genres_count
```

```python
#lets visualize it
genres_count['original_title'].plot.pie(title = 'Movies Per Genre in %',autopct = '%1.1f%%',figsize = (10,10))
```

```python
genres_count['original_title'].plot.barh(title = 'Movies Per Genre' , color = 'DarkBlue',figsize = (10,9))
```

2. Which genres have high avg budget and revenue?

```python
genres_avg = movies_genres.groupby('genres_split').mean(numeric_only = True) #find avg per genres
pd.options.display.float_format = '{:2f}'.format #change the column that are in scientific notation to float

genres_avg
```

```python
genres_avg.sort_values('budget', inplace = True) #sort and save to genres_avg
```

```python
#plot 
genres_avg[['budget','revenue']].plot.barh(title = 'Budget and Revenue by Genre' , color = ('DarkBlue','c'),figsize = (10,9))
```

```python
genres_avg.sort_values('revenue', inplace = True)
```

```python
genres_avg[['budget','revenue']].plot.barh(title = 'Budget and Revenue by Genre' , color = ('DarkBlue','c'),figsize = (10,9))
```

3. Which genres have high avg profit? 

```python
genres_avg
```

```python
genres_avg.sort_values('profit', inplace = True)
```

```python
genres_avg['profit'].plot.barh(title = 'Profit by Genre' , color = 'DarkBlue',figsize = (10,9))
```

4. Which genres have high avg popularity?

```python
genres_avg.sort_values('popularity', inplace = True)
```

```python
genres_avg['popularity'].plot.barh(title = 'Popularity by Genre' , color = 'DarkBlue',figsize = (10,9))
```

5. Which genres have highest number of movies with a voting avg >=8?

```python
movies_genres.head()
```

```python
#to answer question 5 we first will filter on the vote count (becuase if its too low we cant really say much about the voting)
vote_fifty = movies_genres[(movies_genres['vote_count'] >= 50) & (movies_genres['vote_average'] >= 8)]
vote_zero = movies_genres[movies_genres['vote_average'] >= 8] #for good measure
```

```python
genres_vote = vote_fifty.groupby('genres_split')[['original_title']].nunique().sort_values('original_title',ascending = True)
```

```python
genres_vote['original_title'].plot.barh(title = 'Vote Average by Genre number of votes >= 50' , color = 'DarkBlue',figsize = (10,9))
```

```python
genres_vote = vote_zero.groupby('genres_split')[['original_title']].nunique().sort_values('original_title',ascending = True)
```

```python
genres_vote['original_title'].plot.barh(title = 'Vote Average by Genre' , color = 'DarkBlue',figsize = (10,9))
```

5. Look at profit Per Genre Per Year

```python
movies_genres.head()
```

```python
time_genre = movies_genres.groupby(['release_year','genres_split'])[['profit']].mean()
```

```python
time_genre
```

```python
final_genre = pd.pivot_table(time_genre,values = 'profit',index = ['genres_split'],columns = ['release_year'])
```

```python
sns.set(rc = {'figure.figsize' : (15,10)})
sns.heatmap(final_genre,cmap = 'YlGnBu',linewidths = .5)
plt.title('Genres by Profit per Year')
plt.show()
```
